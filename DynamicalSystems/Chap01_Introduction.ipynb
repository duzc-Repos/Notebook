```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
## e.g.1 solve simple ODEs
t = sym.symbols('t')
x = sym.symbols('x', cls=sym.Function)
deqn1 = sym.Eq(x(t).diff(t), 1-x(t))
deqn1
```

```python
soln1 = sym.dsolve(deqn1, x(t))
soln1
```

```python
## e.g.2 second order
t = sym.symbols('t')
y = sym.symbols('y', cls=sym.Function)
deqn2 = sym.Eq(y(t).diff(t, 2) + y(t).diff(t) + y(t), sym.exp(t)  )
deqn2
```

```python
soln2 = sym.dsolve(deqn2)
soln2
```

```python
## surface plot
def flux_qubit_potential(phi_m, phi_p, alpha=0.7, phi_ext=np.pi):
    return 2 + alpha - 2*np.cos(phi_p)*np.cos(phi_m) - alpha*np.cos(phi_ext-2*phi_p)

phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X, Y = np.meshgrid(phi_p, phi_m)
Z = flux_qubit_potential(X, Y).T
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(1, 1, 1, projection='3d')
p = ax.plot_wireframe(X, Y, Z, rstride=4, alpha=0.25)

cset = ax.contour(X,Y,Z,zdir='z', offset=-np.pi, cmap=plt.cm.coolwarm)
cset = ax.contour(X,Y,Z,zdir='x', offset=-np.pi, cmap=plt.cm.coolwarm)
cset = ax.contour(X,Y,Z,zdir='y', offset=3*np.pi, cmap=plt.cm.coolwarm)
ax.set_xlim3d(-np.pi, 2*np.pi)
ax.set_ylim3d(0, 3*np.pi)
ax.set_zlim3d(-np.pi, 2*np.pi)
ax.set_xlabel('$\phi_p$', fontsize=15)
ax.set_ylabel('$\phi_m$', fontsize=15)
ax.set_zlabel('Potential', fontsize=15)
plt.tick_params(labelsize=15)
ax.set_title("Surface and contour plots",fontsize=15)
plt.show()
```

```python
t = np.linspace(-10, 10, 1000)
x = np.sin(t)
y = np.cos(t)
z = t

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(1, 1, 1, projection='3d')
ax.plot(x, y, z)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("3D Parametric Curve")
plt.show()
```
