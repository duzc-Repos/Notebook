```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

# 3.1 Canonical Forms

$\begin{cases} \frac{{\rm d}x_1}{{\rm d}t} = a_{11}x_1 + a_{12}x_2 \\ \frac{{\rm d}x_2}{{\rm d}t} = a_{21}x_1 + a_{22}x_2 \end{cases} \Rightarrow \frac{{\rm d}x}{{\rm d}t} = Ax \Rightarrow \frac{{\rm d}y}{{\rm d}t} = Jy$

where $J = P^{-1}AP$ is one of the following forms:
$$J_1 = \begin{bmatrix} \lambda_1 & 0 \\ 0 & \lambda_2 \end{bmatrix}, J_2 = \begin{bmatrix} \alpha & \beta \\ -\beta & \alpha \end{bmatrix}, J_3 = \begin{bmatrix} \lambda_1 & 0 \\ 0 & \lambda_1 \end{bmatrix}, J_4 =  \begin{bmatrix} \lambda_1 & \mu \\ 0 & \lambda_1 \end{bmatrix}$$


## Definition:

* __trajectory__ or __orbits__: the solution curve $\phi(t) = [x_1(t), x_2(t)]$
    * infinite number of trajectories
    * qualitative behavior can be determined by plotting just a few trajectories
* __phase portrait__: qualitative behavior is determined as $x$ vary with $t$
    * where any trajectory will end up from any given initial condition
* __direction field__ or __vector field__: __gredients__ $\frac{{\rm d}x_2}{{\rm d}x_1}$ and __direction vectors__ of trajectories in phase plane
    * __isoclines__: $\frac{{\rm d}x_2}{{\rm d}x_1}$ is constant
    * __nullclines__: $\frac{{\rm d}x_1}{{\rm d}t} = 0$ and $\frac{{\rm d}x_2}{{\rm d}t} = 0$

## Non-simple Canonical System: $det(A) = 0$
e.g $\frac{{\rm d}x_1}{{\rm d}t} = x_1$, $\frac{{\rm d}x_2}{{\rm d}t} = 0$

Any point lying on the y-axis is the critical points

```python
A = np.array([[1, 0], [0, 0]])
def dx_dt(x, t):
    return np.dot(A, x)

plt.figure(figsize=(10, 10))

ic = np.linspace(-3, 3, 4)
ts = np.linspace(0, 4, 100)
for r in ic:
    for s in ic:
        xs = odeint(dx_dt, [r, s], ts)
        plt.plot(xs[:, 0], xs[:, 1], 'r-')
ts = np.linspace(0, -4, 100)
for r in ic:
    for s in ic:
        xs = odeint(dx_dt, [r, s], ts)
        plt.plot(xs[:, 0], xs[:, 1], 'r-')
 
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.tick_params(labelsize=15)
plt.xlim(-3.5, 3.5)
plt.ylim(-3.5, 3.5)

X, Y = np.mgrid[-3:3:30j, -3:3:30j]
u = A[0, 0] * X + A[0, 1] * Y
v = A[1, 0] * X + A[1, 1] * Y
plt.quiver(X, Y, u, v, color='b')
plt.show()
        
```

```python
# The 2-dimensional linear system.
a, b, c, d = 2, 1, 1, 2
def dx_dt(x, t):
    return [a*x[0] + b*x[1], c*x[0] + d*x[1]]

# Trajectories in forward time.
ts = np.linspace(0, 4, 100)
ic = np.linspace(-1, 1, 5)
for r in ic:
    for s in ic:
        x0 = [r, s]
        xs = odeint(dx_dt, x0, ts)
        plt.scatter(r, s, c='green', s=15)
        plt.plot(xs[:,0], xs[:,1], "r-")

# Trajectories in backward time.
ts = np.linspace(0, -4, 100)
ic = np.linspace(-1, 1, 5)
for r in ic:
    for s in ic:
        x0 = [r, s]
        xs = odeint(dx_dt, x0, ts)
        plt.scatter(r, s, c='green', s=15)
        plt.plot(xs[:,0], xs[:,1], "r-")

# Label the axes and set fontsizes.
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.tick_params(labelsize=15)
plt.xlim(-2, 2)
plt.ylim(-2, 2);
# Plot the vectorfield.
X,Y = np.mgrid[-1:1:10j, -1:1:10j]
u = a*X + b*Y
v = c*X + d*Y
plt.quiver(X, Y, u, v, color = 'b')
plt.show()
```
